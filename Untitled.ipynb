```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
sns.set()
```

```python
X = np.linspace(0,1,100)
```

```python
X.shape
```

```python
exp = np.random.choice([2,3])
```

```python
np.random.randn(X.shape[0])
```

```python
y = X**exp+np.random.randn(X.shape[0])/10
```

```python
plt.plot(X,y,'.')
```

```python
p = np.polyfit(X, y, 1)
z = np.poly1d(p)
```

```python
plt.plot(X,y,'.')
#plt.plot(X,z(X), label=r"Model: ${:.2f}x + {:.2f}$".format.(*p))
plt.plot(X, z(X), label=r"Model: ${:.2f}x + {:.2f}$".format(*p))
#plt.plot(X, X*exp, label = r"Truth: $x")
plt.plot(X, X**exp, label=r'Truth: $x^{}$'.format(exp))
plt.legend()
```
